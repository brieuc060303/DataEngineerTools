```python
import pandas as pd
import pymongo
import json
```

```python
client = pymongo.MongoClient()
database = client['exercices']
collection = database['youtube']
```

```python
df_youtube = pd.read_csv("./data/USvideos.csv")
df_youtube['category_id'] = df_youtube['category_id'].astype(str)
df_youtube.head(2)
```

## Question 0

### Netoyer les données

```python
with open('./data/US_category_id.json', 'r') as json_file:
    categories_json = json.load(json_file)
categories_mapping = {category['id']: category['snippet']['title'] for category in categories_json['items']}
categories_mapping
```

```python
df_youtube['category_name'] = df_youtube['category_id'].map(categories_mapping)
df_youtube.tail()
```

```python
data = df_youtube.to_dict(orient='records')
```

### Importer les données

```python
db_youtube = client.youtube
ytb = db_youtube['apple'] 
ytb.insert_many(data)
```

```python
ytb_all = db_youtube['all']
ytb_all.insert_many(data)
```

## Question 2

```python
unique_categories = ytb.distinct("category_name")
num_unique_categories = len(unique_categories)
num_unique_categories
```

## Question 3

```python
import re
documents = ytb_all.find()

for document in documents:

    tags_list = re.split(r'\|', document['tags'])

    ytb.update_one(
        {"_id": document["_id"]},
        {"$set": {"tags": tags_list}}
    )
```

```python
cur = ytb.find()
next(cur)
```

## Question 4

```python
ytb.create_index([("views", -1)])
```

```python
cur = ytb.find().sort("views", -1)
list(cur)
```

## Question 5

```python
cur = ytb.aggregate([{"$group" : {"_id" : "$category_name", "meanByCat" : {"$avg" : "$views"}}}])
list(cur)
```

## Question 6 

```python
cur = ytb_all.aggregate([{"$group" : {"_id" : "$channel_title", "meanLikes" : {"$avg" : "$likes"}}}])
list(cur)
```

```python
cur = ytb_all.aggregate([
    {"$group": {"_id": "$channel_title", "meanLikes": {"$avg": "$likes"}}},
    {"$sort": {"meanLikes": -1}},
    {"$limit": 10}
])

result = list(cur)
result
```
